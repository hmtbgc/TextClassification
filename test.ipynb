```python
import numpy as np
from tqdm import tqdm
```

```python
path = "./data/pretrained_wordvector/sgns.sogou.char"
```

```python
out = dict()
with open(path, "r") as f:
    for line in tqdm(f):
        x = line.strip().split(' ')
        word = x[0]
        embed = np.asarray(x[1:], dtype=np.float32)
        out[word] = embed
```

```python
len(out)
```

```python
number = 0
for key, value in out.items():
    print(key, value)
    number += 1
    if (number > 10):
        break
```
